## Data Cleaning

```python
dfv3=pd.read_csv("dfv3_test.csv")
dfv3=dfv3.drop(columns=['Unnamed: 0'])
#filter out Price, Bed and Bath na
dfv3=dfv3[dfv3["Bed"].isna()==False]
dfv3=dfv3[dfv3["Bath"].isna()==False]
dfv3=dfv3[dfv3["Price"].isna()==False]
dfv3.isna().sum()
```

We will fill Year built missing value with mode based on city.

```python
mode=dfv3[dfv3["City"]=='SanJose']['Year built'].mode()[0]
print(mode)
dfv3['Year built'].fillna(mode,inplace = True)
dfv3['Year built'].isna().sum()
```

We will fill missing value in lot size with median.

```python
#Lot size(sqft):median
median=dfv3['Lot size(sqft)'].median()
print(median)
dfv3['Lot size(sqft)'].fillna(median,inplace = True)
dfv3['Lot size(sqft)'].isna().sum()
```

For Heating columns, we will first fill missing values with mode, then transform it to a boolean column.

```python
heat_mode=dfv3['Heating'].mode()[0]
print('Mode of heating column is', heat_mode)
dfv3['Heating'].fillna(heat_mode,inplace = True)
dfv3['Heating'].isna().sum()
dfv3['Heating']=pd.Series(dfv3['Heating'], dtype="string")
heating_list=dfv3['Heating'].tolist()
heating_bool=[False if 'none' in i.lower() else True for i in heating_list]
dfv3['Heating_bool']=pd.Series(heating_bool)
print(dfv3['Heating'].head())
dfv3['Heating_bool'].head()
```

For Cooling columns, we will also fill missing values with mode, then transform it to a boolean column.

```python
cool_mode=dfv3['Cooling'].mode()[0]
dfv3['Cooling'].fillna(cool_mode,inplace = True)
print(dfv3['Cooling'].value_counts())
#Transform to boolean value
cooling_list=dfv3['Cooling'].tolist()
cooling_bool=[False if i.lower()=='none' else True for i in cooling_list]
dfv3['Cooling_bool']=pd.Series(cooling_bool)
print(dfv3['Cooling_bool'].value_counts())
```

```python
dfv3.drop(columns=['Cooling', 'Heating'],inplace=True)
```

We will replace missing value in HOA to 0.

```python
dfv3['HOA']= dfv3['HOA'].fillna(0)
```

We will fill missing value in parking type using mode based on city, and 1 for parking space.

```python
#SanJose Parking Type
SJMode = dfv3[dfv3['City']== 'SanJose']['Parking type'].mode()[0]
print(SJMode)
dfv3.loc[dfv3['City'] == 'SanJose']['Parking type'].fillna(SJMode, inplace = True)
#Richmond Parking Type
RichMode = dfv3[dfv3['City']== 'Richmond']['Parking type'].mode()[0]
print(RichMode)
dfv3.loc[dfv3['City'] == 'Richmond']['Parking type'].fillna(RichMode, inplace = True)
#SanLeandro Parking Type
SLMode = dfv3[dfv3['City']== 'SanLeandro']['Parking type'].mode()[0]
print(SLMode)
dfv3.loc[dfv3['City'] == 'SanLeandro']['Parking type'].fillna(SLMode, inplace = True)
#SanMateo Parking Type
SMMode = dfv3[dfv3['City']== 'SanMateo']['Parking type'].mode()[0]
print(SMMode)
dfv3.loc[dfv3['City'] == 'SanMateo']['Parking type'].fillna(SMMode, inplace = True)
dfv3['Parking space'].fillna(1, inplace = True)
```

```python
dfv3['Sold date'] = dfv3['Sold date'].replace(['99/99/9999'], '01/01/1990')
dfv3['Sold date'] = pd.to_datetime(dfv3['Sold date'], format = '%m/%d/%y',infer_datetime_format=True)
```
